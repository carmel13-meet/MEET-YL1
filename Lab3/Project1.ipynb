```python
import csv
f = open('crime.csv', 'rb')
r = csv.reader(f)

year = []
crimes = []

for row in r:
    x = row[0].replace("\t", " ")
    z = x.split()
    z = z[7:]
    s = map(int, z)
    year.append(s[0])
    crimes.append(s[1])
f.close()
```

```python
from pylab import *
```

My project shows how many crimes were made in Alabama during the specified years

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(year, crimes)
ax.set_title("Year and No. of Crimes")
ax.set_xlabel('Year')
ax.set_ylabel('No. of Crimes')
savefig("IndividualProjectEASIER.png")
```

As you can see, there is no unequivocal conclusion with which we can assume the connection between the year and the number of crimes that happened. But still, the graph looks fine :-)
