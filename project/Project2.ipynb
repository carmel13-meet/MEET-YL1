```python
import csv 
f = open('motor_vehicles.csv', 'rb')
r = csv.reader(f)
i = 0
myList = []
myList2 = []
for row in r:
    i = i + 1
    if i == 1:
        myList = row
    if i == 2:
        myList2 = row
myList = myList[4: 57]
myList = map(float, myList)
myList = myList[49: 53]
myList2 = map(float, myList2[49: 53])
f.close()
```

My project shows the number of motor vehicles for every 1000 people in Afghanistan from 2010 to 2013

```python
from pylab import *
fig = plt.figure()
ax = fig.add_subplot(111)
ax.bar(myList, myList2)
ax.set_title("Graph in Afghanistan")
ax.set_xlabel('Years')
ax.set_ylabel('Number of motor vehicles for every 1000 people')
savefig("IndividualProjectHARDER.png")
```

As you can see, when the years went on, the graphs went up :-)
